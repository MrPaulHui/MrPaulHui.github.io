```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
from matplotlib.colors import rgb2hex
```

```python
travelled_place = {'江苏省', '上海市', '湖南省', '浙江省', '湖北省', '四川省', '重庆市', '河南省', '陕西省', '江西省', '山东省', '安徽省'}
```

```python
cmp= Basemap(llcrnrlon=80, llcrnrlat=0, urcrnrlon=140, urcrnrlat=51, projection='lcc', lat_1=33, lat_2=45, lon_0=100)
cmp.readshapefile("china-shapefiles-master/shapefiles/china",  'china', drawbounds=True)
cmp.readshapefile("china-shapefiles-master/shapefiles/china_nine_dotted_line",  'nine_dotted', drawbounds=True)
color = rgb2hex((0.8633141099577085, 0.0835678585159554, 0.11687812379853903))
province_dict = {}
for i in range(len(cmp.china)):
    key = cmp.china_info[i]['OWNER'].replace("\x00","")
    if key not in province_dict:
        province_dict[key] = [cmp.china[i], 0]
for key in province_dict:
    if key in travelled_place:
        province_dict[key][1] = 1
ax = plt.gca()
for province_name, value in province_dict.items():
    #print(province_name)
    #print(value[0])
    poly = Polygon(value[0], facecolor=color, edgecolor=color)
    ax.add_patch(poly)
plt.show()
```
